# 3 - U-Net Model


UNet has proven to be effective for a variety of image segmentation tasks, such as medical image segmentation (e.g., tumor segmentation, cell segmentation), satelite imagery segmentation and other general image segmentation applications. Its architecture's ability to capture fine-grained details and contextual information has made it widely adopted and serves as the basis for many subsequent segmentation models.

```python
# importting Unet model
from UNet import UNet
from tensorflow.keras.optimizers import Adam

model = UNet(128, 128).model()
model.compile(optimizer=Adam(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
# model.summary()
```

<a name='4'></a>
# 4 - Training 

```python
num_train_imgs = 552  
num_val_imgs = 138 
batch_size = 16
train_steps_per_epoch = num_train_imgs//batch_size
val_steps_per_epoch = num_val_imgs//batch_size

history=model.fit(train_image_gen,
                  steps_per_epoch=train_steps_per_epoch,
                  validation_data=val_image_gen,
                  validation_steps=val_steps_per_epoch,
                  epochs=5,
                  batch_size=batch_size)

model.save('model.h5')
```

```python
# Plot the training and validation accuracy and loss at each epoch
loss = history.history['loss']
val_loss = history.history['val_loss']
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
epochs = range(1, len(loss) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(epochs, loss, 'y', label='Training loss')
ax1.plot(epochs, val_loss, 'r', label='Validation loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(epochs, np.array(train_accuracy) * 100, label='Training accuracy')
ax2.plot(epochs, np.array(val_accuracy) * 100, label='Validation accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.legend()

plt.suptitle('Training and validation loss and accuracy')
plt.show()
```

# Making Predictions

```python
# Loading pretrained model
from keras.models import load_model
model = load_model('model.h5')
```

```python
# Visualizing the test image
test_file = 'test/test.tif'
test_img = cv2.imread(test_file)
plt.imshow(test_img*255)
```

```python
# Dividing test image into patches

test_image_patches = tf.image.extract_patches(images=tf.expand_dims(test_img, 0),
                                   sizes=[1, 128, 128, 1],
                                   strides=[1, 128, 128, 1],
                                   rates=[1, 1, 1, 1],
                                   padding='SAME')

patched_images = tf.reshape(test_image_patches, shape=(test_image_patches.shape[1]*test_image_patches.shape[2], 128, 128, 3)).numpy().astype("uint8")
```

```python
# Visualizing patched test image
plt.figure(figsize=(10, 10))

for imgs in test_image_patches:
    print(imgs.shape)
    count = 0
    for r in range(imgs.shape[0]):
        for c in range(imgs.shape[1]):
            ax = plt.subplot(imgs.shape[0], imgs.shape[1], count+1)
            a = tf.reshape(imgs[r,c]*255,shape=(128,128,3)).numpy().astype("uint8")
            ax.imshow(a)
            count += 1
```

```python
# Predicting the mask
predictions = model.predict(patched_images)
test_pred_batch_argmax = np.argmax(predictions, axis=3)
```

```python
# Visualizing predicted mask
image_array = test_pred_batch_argmax  # Your array of images

num_rows = 4
num_cols = 7

# Create a figure and axis with the desired aspect ratio
fig, ax = plt.subplots(num_rows, num_cols, figsize=(num_cols * 1.5, num_rows * 1.5))

# Flatten the axis array to simplify indexing
ax = ax.flatten()

# Iterate over the images and plot each image
for i in range(len(image_array)):
    image = image_array[i]
    ax[i].imshow(image, cmap='gray',vmin =0,vmax = 1)  # Assuming the images are grayscale, change cmap if needed
    ax[i].axis('off')

# Remove extra subplots if there are fewer images than the total number of subplots
if len(image_array) < num_rows * num_cols:
    for j in range(len(image_array), num_rows * num_cols):
        fig.delaxes(ax[j])

# Adjust the layout and spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Merging the patches
test_pred_batch_argmax = np.reshape(test_pred_batch_argmax, (test_image_patches.shape[1], test_image_patches.shape[2], 128, 128))
test_pred_batch_argmax = np.transpose(test_pred_batch_argmax, (0, 2, 1, 3))
test_pred_batch_argmax = np.reshape(test_pred_batch_argmax, (test_image_patches.shape[1]*128, test_image_patches.shape[2]*128))

# Visualizing the predicted mask   
plt.imshow(test_pred_batch_argmax, cmap='gray',vmin =0,vmax = 1)
```
